### Pandas Data Pre-processing
Now we will process our data so it is ready use for statsmodels and matplotlib visualisation.

### Processing microplastics dataset

```python
# Select only microplastics from 1950 to 2019
# Select one of the categories => slice from 1950 to 2019
micro_p_19502019 = micro_p[micro_p['Entity'] == 'Emissions growth to 2050'].set_index('Year').loc[1950:2019]

# Rename column and select only macroplastics tonnes
micro_p_19502019_renamed = micro_p_19502019.rename(columns={
    "Accumulated ocean plastic: Microplastics (<0.5cm)": "microplastics_tonnes"})[['microplastics_tonnes']]

# convert tonnes to hundred thousand tonnes for easier viewing of numbers
micro_p_19502019_renamed['microplastics_tonnes'] = micro_p_19502019_renamed['microplastics_tonnes'] / 100000
micro_p_renamed_ht_tonnes = micro_p_19502019_renamed.rename(columns={'microplastics_tonnes': 'microplastics_ht_tonnes'})

print(micro_p_renamed_ht_tonnes.head())
print(micro_p_renamed_ht_tonnes.tail())
```

### Processing plastic production dataset

From earlier, plastic production has 69 rows, but microplastics have 70 rows. This means that one of the years for plastic production is missing. We shall fill in the missing year using forward fill, since this is time-series data. 

Plastic production of current year may be dependent on plastic production of the past year, so it would be more accurate to fill the value based on the previous year value.

```python
# Select only plastic production and year from p_prod
p_prod = p_prod[['Year', 'Annual plastic production between 1950 and 2019']]
# Rename column
p_prod_renamed = p_prod.rename(columns={'Annual plastic production between 1950 and 2019': 'plastic_prod_tonnes'})

# iterate through p_prod year with a complete range of years from 1950 to 2019
all_years = list(range(1950, 2020))
missing_year = 0
for i in range(70):
    if p_prod['Year'].iloc[i] != all_years[i]:
        # If the year is different, it means that p_prod has this missing year.
        print(f"Missing year is {all_years[i]}.")
        missing_year = all_years[i]
        break

# insert the missing year
p_prod_renamed.loc[len(p_prod_renamed.index)] = [missing_year, np.NaN] 

# print number of non-null rows before filling
print(f"Before filling: {p_prod_renamed['plastic_prod_tonnes'].notnull().sum()} non-missing rows")

# sort by year, then fill missing year
p_prod_renamed = p_prod_renamed.sort_values('Year').fillna(method='ffill')

# print number of non-null rows after filling
print(f"After filling: {p_prod_renamed['plastic_prod_tonnes'].notnull().sum()} non-missing rows")

# set index, slice 2 years before and after the missing year to show the filling
p_prod_renamed = p_prod_renamed.set_index('Year')
print(p_prod_renamed.loc[(missing_year -2):(missing_year +2)])

# convert tonnes to hundred thousand tonnes for easier viewing of numbers
p_prod_renamed['plastic_prod_tonnes'] = p_prod_renamed['plastic_prod_tonnes'] / 100000
p_prod_renamed_ht_tonnes = p_prod_renamed.rename(columns={'plastic_prod_tonnes': 'plastic_prod_ht_tonnes'})

print('\nCurrent Dataset:')
print(p_prod_renamed_ht_tonnes.head())
print(p_prod_renamed_ht_tonnes.tail())
```

### Combine the 2 datasets

We shall combine the 2 datasets together for convenience.

```python
# use join method to join dataframes at index (year)
df1i = p_prod_renamed_ht_tonnes.join(micro_p_renamed_ht_tonnes)
df1i.head()
```

```python
df1i.tail()
```

### Applying Statsmodels linear regression analysis
Now that our dataframe is ready, we can conduct Statsmodels regression analysis to find out more about the relationship between plastic production and microplastics in ocean.

### Simple linear regression: Global plastic production (hundred thousand tonnes) vs Microplastics (hundred thousand tonnes) in ocean

```python
# Dependent variable microplastics_ht_tonnes vs independent variable plastic_prod_ht_tonnes
linearRegModel = smf.ols("microplastics_ht_tonnes ~ plastic_prod_ht_tonnes", data=df1i).fit()

print(linearRegModel.params)
print(linearRegModel.summary())
```

Insights

```python
# Get x and y values
x=df1i['plastic_prod_ht_tonnes']
y=df1i['microplastics_ht_tonnes']

# define figure and axes
plt.figure(figsize=(9, 7))
ax = plt.axes()
ax.set_facecolor("aliceblue")

# Plot the scatterplot for y vs x
plt.scatter(y=y, x=x, marker='1', s=200, c='forestgreen')

# best fit line Y = C + M * X
m,c = np.polyfit(x, y, deg=1)

plt.plot(x, c+m*x, 'b-')

plt.xlabel('Global Plastic Production (hundred thousand tonnes)')
plt.ylabel('Microplastics in Surface Ocean (hundred thousand tonnes)')

plt.xticks(np.linspace(0, 5000, num=11))
plt.yticks(np.linspace(0, 6, num=13))

plt.xlim([0, 5000])
plt.ylim([0, 6.0])

plt.grid()

plt.title('Scatterplot of Microplastics in surface ocean vs Global plastic production ')

# display the equation of the Single linear Regression Model
linearModEq = f"y = {linearRegModel.params[0]:.4} + {linearRegModel.params[1]:.3}x"

plt.annotate(linearModEq, xy=(1500, 3.75), xytext=(1500,3.75), size=12, bbox={'facecolor':'greenyellow'})

# Predict y when x = 3000
testingValue={'plastic_prod_ht_tonnes': 3000}
predictedValue = linearRegModel.predict(testingValue)

# point on the linear equation when x = 3000
plt.plot(3000, predictedValue, 'ro', markersize=10)

desc = f'x = 3000 hundred tonnes,\ny = {predictedValue[0]:.2f} hundred tonnes'
ax.annotate(desc, xy=(3050, 2.9), xytext=(3500, 2), arrowprops=dict(facecolor='red'), bbox={'facecolor':'greenyellow'})

plt.show()
```

### Insights:

The R-squared is 0.955, which is very close to 1. This means that the linear equation fits the data very well!
This model explains 95.5% of the variation of Accumulated microplastics in the surface ocean.

As Global Plastic Production increases, Microplastics in Surface Ocean increases.

Sub x = 3000 in the equation:

### Pandas Data Pre-processing
Now we will process our data so it is ready for matplotlib visualisation.

```python
# remove rows like Asia, Africa and Oceania which are geographical regions/continents, not countries, and do not have a Code.
# since the only column with null values is Code and they all have null for Code
df1ii = df1ii.dropna()

# drop the 'World' row, since we want to see countries/territories
df1ii = df1ii.drop((df1ii[df1ii['Entity'] == 'World']).index)

# rename column for convenience
df1ii = df1ii.rename(columns={'Mismanaged plastic waste per capita (kg per year)': 'plastic_per_capita'})

# verify that null rows have been dropped
df1ii.info()
```

Apply user-defined function to check if the row plastic_per_capita is an upper outlier

```python
# Get the upper fence
quartiles = df1ii['plastic_per_capita'].quantile([0.25, 0.75])
upper_fence = quartiles[0.75] + 1.5 * (quartiles[0.75] - quartiles[0.25])

# Define user-defined function to check for upper outlier
def is_upper_outlier(row):
    if row['plastic_per_capita'] > upper_fence:
        return 'Upper Outlier'
    else:
        return 'Not Outlier'

# Apply the user-defined function to a new column in the DataFrame
df1ii['is_upper_outlier'] = df1ii.apply(is_upper_outlier, axis=1)
```

Splitting outlier and not outlier data so that separate histograms can be plotted on the same axes. This allows me to give different colours to separate them visually. 

```python
# Subset dataframe into Upper Outlier and Not Outlier
is_outlier = df1ii[df1ii['is_upper_outlier'] == 'Upper Outlier'].sort_values('plastic_per_capita')
not_outlier = df1ii[df1ii['is_upper_outlier'] == 'Not Outlier'].sort_values('plastic_per_capita')
print(f'Not Outlier:\n{not_outlier.head()}\n')
print(f'Upper Outlier:\n{is_outlier}\n')

# Create separate mismanaged plastic waste per capita of the whole dataset subset for easier access and readability of code
mismanaged_plastic = df1ii['plastic_per_capita'].sort_values()
print(f'Selecting only mismanaged plastic per capita:\n{mismanaged_plastic.head()}')
```

### Visualisation with matplotlib

```python
# Create subplots
fig, ax = plt.subplots(2, 2, figsize=(12, 9))

# Set figure title
fig.suptitle('Distribution of mismanaged plastic waste per capita (kg) by country in 2019')

# Boxplot on top left
# Define boxplot styles
boxprops = dict(linestyle='-', linewidth=1.5, color='forestgreen')
flierprops = dict(marker='o', markerfacecolor='lightcoral', markersize=10)
medianprops = dict(linewidth=3, color='forestgreen')
ax[0, 0].boxplot(mismanaged_plastic, vert=False, boxprops=boxprops, flierprops=flierprops, medianprops=medianprops)
ax[0, 0].set_title('Boxplot View')
ax[0, 0].set_xticks(np.linspace(0, 80, num=17))
ax[0, 0].set_xlabel('Mismanaged plastic waste per capita (kg)')
ax[0, 0].set_ylabel('Number of countries')

# Histogram on bottom left
ax[1, 0].hist(is_outlier['plastic_per_capita'], color='lightcoral', bins=10, label='Upper Outlier')
ax[1, 0].hist(not_outlier['plastic_per_capita'], color='forestgreen', bins=10, label='Not Outlier')
ax[1, 0].set_title('Histogram View')
ax[1, 0].set_yticks(np.linspace(0, 80, num=17))
ax[1, 0].set_xticks(np.linspace(0, 80, num=17))
ax[1, 0].set_xlabel('Mismanaged plastic waste per capita (kg)')
ax[1, 0].set_ylabel('Number of countries')

# Bar graph on the top right
ax[0, 1].bar(is_outlier['Entity'], is_outlier['plastic_per_capita'], color='lightcoral')
plt.setp(ax[0, 1].get_xticklabels(), rotation=45, ha="center")
ax[0, 1].set_title('Viewing upper outliers')
ax[0, 1].set_xlabel('Countries')
ax[0, 1].set_ylabel('Mismanaged plastic waste per capita (kg)')

# Normal distribution qq plot on the bottom right
sm.qqplot(mismanaged_plastic, line='s', ax=ax[1, 1])
ax[1, 1].set_yticks(np.linspace(-20, 80, num=11))
ax[1, 1].set_xticks(np.linspace(-3, 3, num=7))
ax[1, 1].set_title('QQ Plot (Normal Distribution)')

# Adjust layout for better spacing
plt.tight_layout()

# Show subplots
plt.show()
```

### Understanding methods of plastic waste management
reference: https://enterclimate.com/blog/methods-of-plastic-waste-management/

Recycled: Plastic that is recycled, put to other uses. used plastic can be made into a reusable form, which is more sustainable since less plastic has to be produced. Lesser use of resources and harmful byproducts like carbon dioxide.

Incinerated: Plastic that is burnt. This can provide energy for other uses at the same time. however, Harmful gases will be released that have to be treated. This method also emits carbon dioxide to the environment which may accelerate global warming. This requires less land since only the incineration plant is required.

Landfilled: Plastic that is buried underneath land. Land that is designated as landfills may not be available for other uses such as agriculture (farming), so it is like a 'waste of space'. The landfill aim is to safeguard the environment from the plastic waste. Basically wait for the plastic waste to decompose. Plastics only moderately degrade so the land will not be usable for a very long time.

We shall compare the municipal waste recycling rates of:
- USA in North America,
- 2 countries in Europe: Germany and the United Kingdom (UK)
- 1 country in Asia: Korea,

From 2005 to 2014.

### Pandas pre-processing

```python
# rename column
df2ii = df2ii.rename(columns={'Variable:% Recycling - MUNW': 'mun_waste_recycling_rate'})

# countries to subset
countries_list = ['United States', 'United Kingdom', 'Korea', 'Germany']
# range of years from 2005 to 2014. the respective mun_waste_recycling_rate will be merged to this year column
countries_df = pd.DataFrame({'Year': range(2005, 2015)})

# loop through the 4 countries
for country in countries_list:
    # subset the country
    subset = df2ii[df2ii['Entity'] == country][['mun_waste_recycling_rate', 'Year']]
    # rename the column to the country name for unique identifier of column
    subset = subset.rename(columns={'mun_waste_recycling_rate': country})
    # merge, maintaining the year range of 2005 to 2014
    countries_df = pd.merge(countries_df, subset, on='Year', how='left')

# show resulting dataframe
countries_df
```

We have NaN missing values for Korea from 2006 to 2007. We shall use linear interpolation to fill the missing values, since we know the values for the year before and after.

```python
# linear (default) interpolation
countries_df = countries_df.interpolate()

# Verify that missing values have been interpolated
countries_df
```
